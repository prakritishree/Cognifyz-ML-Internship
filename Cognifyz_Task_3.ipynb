1) Loading dataset and importing libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

res_data = pd.read_csv('Dataset .csv')
print(res_data.head())
```

```python
res_data.columns
```

2)data cleaning

```python
res_data=res_data.drop(columns=['Restaurant ID','Country Code','City','Address','Locality','Locality Verbose','Longitude','Latitude','Average Cost for two','Currency','Has Table booking','Has Online delivery','Is delivering now','Switch to order menu','Price range','Aggregate rating','Rating color','Rating text','Votes'])
res_data
```

```python
#handling missing values
res_data.isnull().sum()
```

```python
res_data = res_data.dropna(subset=['Cuisines'])
```

3)Data Preprocessing

```python
#Dealing with multiple values in cuisines
res_data['Cuisines'] = res_data['Cuisines'].str.split(', ')
res_data=res_data.explode('Cuisines')
res_data.head()
```

4)Data Spliting

```python
# Encode the 'Cuisines' labels
le = LabelEncoder()
res_data['Cuisines'] = le.fit_transform(res_data['Cuisines'])

# Prepare features using TfidfVectorizer on 'Restaurant Name'
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(res_data['Restaurant Name'])

# Target variable
y = res_data['Cuisines']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

5)Model Training and observation

**1) LOGISTIC REGRESSION MODEL**

```python
# Initialize and train the Logistic Regression model
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)

# Make predictions with Logistic Regression
lr_y_pred = lr_model.predict(X_test)

# Evaluate Logistic Regression model
lr_accuracy = accuracy_score(y_test, lr_y_pred)
print(f'Logistic Regression Accuracy: {lr_accuracy:.2f}')
print('Logistic Regression Confusion Matrix:')
print(confusion_matrix(y_test, lr_y_pred))
```

**2) RANDOM FOREST CLASSIFIER MODEL**

```python
# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions with Random Forest
rf_y_pred = rf_model.predict(X_test)

# Evaluate Random Forest model
rf_accuracy = accuracy_score(y_test, rf_y_pred)
print(f'Random Forest Accuracy: {rf_accuracy:.2f}')
print('Random Forest Confusion Matrix:')
print(confusion_matrix(y_test, rf_y_pred))
```

**3)SUPPORT VECTOR MACHINE MODEL**

```python
# Initialize and train the Support Vector Machine model
svm_model = SVC(kernel='linear', random_state=40)
svm_model.fit(X_train, y_train)

# Make predictions with Support Vector Machine
svm_y_pred = svm_model.predict(X_test)

# Evaluate Support Vector Machine model
svm_accuracy = accuracy_score(y_test, svm_y_pred)
print(f'Support Vector Machine Accuracy: {svm_accuracy:.2f}')
print('Support Vector Machine Confusion Matrix:')
print(confusion_matrix(y_test, svm_y_pred))
```

**4) DECISION TREE CLASSIFIER MODEL**

```python
# Initialize and train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=40)
dt_model.fit(X_train, y_train)

# Make predictions with Decision Tree
dt_y_pred = dt_model.predict(X_test)

# Evaluate Decision Tree model
dt_accuracy = accuracy_score(y_test, dt_y_pred)
print(f'Decision Tree Accuracy: {dt_accuracy:.2f}')
print('Decision Tree Confusion Matrix:')
print(confusion_matrix(y_test, dt_y_pred))
```

**CONCLUSION :

The best model is SVM with the highest accurace of 0.25, followed by Logistic Regression having 0.24 accuracy.

Decision Tree Classifier (Accuracy : 0.14) and Random Forest Classifier (Accuracy : 0.16) performed poorly for the purpose**
